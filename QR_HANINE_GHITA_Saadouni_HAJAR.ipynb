```python
# code QR

!pip install qrcode[pil]

import qrcode

url = "https://colab.research.google.com/drive/1KPxIZm0BRgTSxkCyTtpKdXTBoNsD-D5R?authuser=0#scrollTo=jStqbiYb7l9v;"

qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)
qr.add_data(url)
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="white")
img.save("qrcode_colab.png")

from IPython.display import Image
Image("qrcode_colab.png")
```
